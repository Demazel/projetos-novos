# Correções

```python
# customers - age - valor ausente
media_idade = df_customer['age'].mean()
df_customer['age'] = df_customer['age'].fillna(media_idade)
display(df_customer)
```

```python
# rentals - daily_rate cpm valor < 0
df_rentals.loc[df_rentals['daily_rate']<=0, 'daily_rate'] = pd.NA
display(df_rentals)
```

```python
# vehicles - co2_g_per_km - null
df_vehicles['co2_g_per_km'] = df_vehicles['co2_g_per_km'].fillna(120)
display(df_vehicles)
```

```python
# rentals - daily_rate esta como string
df_rentals['daily_rate'] = pd.to_numeric(df_rentals['daily_rate'], errors='coerce')
display(df_rentals)
```

```python
# rentals - data invalida
df_rentals['start_date'] = pd.to_datetime(df_rentals['start_date'], errors='coerce')
df_rentals['end_date'] = pd.to_datetime(df_rentals['end_date'], errors='coerce')
display(df_rentals)
```

# Engenharia de Dados

```python
# criar backup de dados
df_carbon_clean = df_carbon
df_customer_clean = df_customer
df_rentals_clean = df_rentals
df_vehicles_clean = df_vehicles
```

```python
df_rentals_clean['duração_locação'] = (df_rentals_clean['end_date'] - df_rentals_clean['start_date']).dt.days
display(df_rentals_clean)
```

```python
# receita total por locacao
df_rentals_clean['receita_total'] = df_rentals_clean['daily_rate'] * df_rentals_clean['duração_locação']
display(df_rentals_clean)
```

```python
# faça merge com a base de dados de veiculo
df_rental_vehicles = df_rentals_clean.merge(df_vehicles_clean, on='vehicle_id')
df_rental_vehicles = df_rental_vehicles.merge(df_customer_clean, on='customer_id')
display(df_rental_vehicles)
```

```python
# calcule a emissao de co2 por locacao
df_rental_vehicles['emissao'] = df_rental_vehicles['co2_g_per_km'] * df_rental_vehicles['distance_km']
display(df_rental_vehicles)
```

```python
# Extraia ano e mês da data de início da locação.
df_rental_vehicles['mes'] = (df_rental_vehicles['start_date']).dt.month
df_rental_vehicles['ano'] = (df_rental_vehicles['start_date']).dt.year
display(df_rental_vehicles)
```
